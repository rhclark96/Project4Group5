```python
#Importing libraries
import pandas as pd
from pathlib import Path
```

```python
#Read csv file into a DataFrame
parkinsons_data = Path("../Resources/parkinsons_disease_data.csv")
data_df = pd.read_csv(parkinsons_data)

data_df
```

```python
#See all columns to determine if any are unnecessary, can be dropped
data_df.columns
```

```python
#Drop 'PatientID and 'DoctorInCharge' columns
data_df.drop(["PatientID", "DoctorInCharge"], axis=1, inplace=True)
data_df.columns
```

```python
#Drop any rows with missing values
data_df.dropna(axis=0, how='any')
data_df
```

```python
#Save cleaned DataFrame to CSV
data_df.to_csv("Cleaned_PD_Data.csv")
```
